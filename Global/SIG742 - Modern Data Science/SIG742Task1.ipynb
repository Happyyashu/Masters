ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31] and the length of the list is 21<BR>
• find the average value of age (don’t use numpy and do not import other module or library, do not
directly use math,statistics module functions except the function sum);<br>
• find the standard deviation from the age list <br>


```python
# Creating a function to calculate mean of numbers
def calculate_mean(list_num):
    sum_of_num = 0
    count_of_num = 0

    # Iterating each number from the list of numbers
    for number in list_num:

        # Adding each number to initial sum
        sum_of_num = sum_of_num + number

        # Counting the number of elements in a list
        count_of_num = count_of_num + 1

    # Returning the mean = summation of observations /  number of observations
    return sum_of_num / count_of_num
```

```python
# Creating a function to calculate standard deviation of numbers
def calculate_std(mean_value,list_num):
    squared_sum = 0
    count_of_num = 0
    for n in list_num:
        squared_sum = squared_sum + (n - mean_value)**2
        count_of_num = count_of_num + 1
    return (squared_sum / count_of_num)**(1/2)
```

```python
# Input to the above functions to calculate mean & standard deviation
ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]
```

```python
mean_ages = calculate_mean(ages)
std_ages = calculate_std(mean_ages,ages)
print(f"average value of age: {mean_ages} \nstandard deviation from the age:{std_ages}")
```

Writing the code or function to achieve below requirements:<br>
• You are given the heads of two sorted linked lists list1 = [1,2,4] and<br> list2 = [1,3,4]. Merge
the two lists in a one sorted list which is [1,1,2,3,4,4]. <br>
module / library);<br>
• You are given the heads of two sorted linked lists list1 = [1,2,4] and list2 = [1,3,4].<br> Merge
the two lists in a one sorted list without duplicated elements which is [1,2,3,4].

```python
def one_sorted_list(list_1,list_2):
    """Function takes 2 lists as arguments and returns the merged sorted list and merged sorted list with no duplicate values"""
    list_1.extend(list_2)
    list1 = sorted(list_1)
    sorted_list = list1
    list_no_duplicate = list(set(list1))
    return sorted_list,list_no_duplicate
```

```python
list1 = [1,2,4]
list2 = [1,3,4]
print(f"List 1:{list1}")
print(f"List 2:{list2}")
sorted_list,list_no_duplicate = one_sorted_list(list1,list2)

print(f"Sorted merged list:{sorted_list}")
print(f"Non-dulicated values in list:{list_no_duplicate}")
```

writing a function to print 1,2,3,4,5,x1,6,1,2,3,sig742
(must define a function to run, the input for function is list A, run the function and print the results
line by line or in one line)


```python
# empty list to store the elements from the given list,which will be unpacked
unpack_list = []

def unpack_list_elements(l1st_1):
    """This is a recursive function which is called inside this function again if the element in a list is an instance of list class """
    for a in l1st_1: # Iterate each element in a list and check if the element is of instance of list class
        if isinstance(a,list) == False:
            unpack_list.append(a)
        else:
            unpack_list_elements(a)
```

```python
list_A = [1,2,3,[4,5,['x1',6,[1,2,3,'sig742']]]]
print(f"Given list with elements: {list_A}")
unpack_list_elements(list_A)
print(f"\nlist with unpacked elemts: {unpack_list}")
```

```python
def sum_digits(student_id):
    """Function takes student id as an argument and returns the sum of the digits in student_id """
    sum_of_digits = 0
    for digit in student_id:
        if digit.isdigit(): # to check if each character is a digit or not
            sum_of_digits = sum_of_digits + int(digit)
    return sum_of_digits  # returning the sum of the digits in the student_id

def check_studentid(result):
    """Function takes the argument returned by sum_digits function and returns even/odd based on the argument"""
    if result % 2 == 0:
        return 'even'
    else:
        return 'odd'
```

Given an array of integers [1, 2, 6, 8], return indices of the two numbers such that they add up
to 10.You may assume that each input would have exactly one solution.
Explanation
Given nums = [1, 2, 6, 8], target = 10,
Because nums[1] + nums[3] = 2 + 8 = 10,
return [1, 3].
Your code should have a defined function which could take the array nums= [1, 2, 6, 8] as input
and run the function to return a list [1,3] (only [1,3])


```python
def indices_sum_10(li):
    """Function takes a list values as arguments and return the index of values whose sum add to 10"""
    for num in li:
        a = num
        index = li.index(num)
        for num2 in li[index+1:]:
            s = a + num2
            if s == 10:
                return li.index(a),li.index(num2)
```

```python
list_1 = [1,2,6,8]
list_index = list(indices_sum_10(list_1))
print(f"Indices whose values are added to the value '10' in a list{list_1}: {list_index}")
```

```python
list_2 = [3,1,7,5]
list_index = list(indices_sum_10(list_2))
print(f"Indices whose values are added to the value '10' in a list{list_2}: {list_index}")
```

You will need to write a code to find out what is the probability to get the car if you switch the door?<br>
One possible solution You first want to define a variable with random choice to indicate

```python
# Function to evaluate the probabilty if the user does not want to switch the door

def no_switch_doors():
    number_of_iterations = range(1,1000)
    np.random.seed(5)
    no_switch_doors = []
    wins = 0
    losses = 0
    for n in tqdm(number_of_iterations):
        car_variable_door = np.random.randint(1,4)
        challenger_door = np.random.randint(1,4)
        host = random.choice([i for i in range(1,4) if i not in [car_variable_door,challenger_door]])
        # print()
        # print(car_variable_door,challenger_door,host)
        if challenger_door == car_variable_door:
            wins = wins + 1
        else:
            losses = losses + 1

        win_perc = wins / (wins + losses)
        no_switch_doors.append(win_perc)
        # print(f"\nProbabilty if the User does not want to switch the door running {n} times:--> {win_perc}")
    return (win_perc),no_switch_doors,number_of_iterations
```

```python
result_no_switch,no_switch_doors,number_of_iterations = no_switch_doors()
print(f"\n\nProbabilty if the User does not want to switch the door:--> {result_no_switch}")
```

```python
plt.figure(figsize = (20,6))
plt.scatter(x = number_of_iterations,y=no_switch_doors,alpha=0.4)
plt.xlabel("Number of iterations")
plt.ylabel("Probability of winning the car")
plt.title("Visualizing the probability values of winning the car if user does not switch the door");
```

```python
# Function to evaluate the probabilty if the user wants to switch the door
def switch_doors():
    number_of_iterations = range(1,1000)
    np.random.seed(5)
    switch_doors = []
    wins = 0
    losses = 0
    for n in tqdm(number_of_iterations):
        car_variable_door = np.random.randint(1,4)
        challenger_door = np.random.randint(1,4)
        host = random.choice([i for i in range(1,4) if i not in [car_variable_door,challenger_door]])
        challenger_door_changed = random.choice([i for i in range(1,4) if i not in [host,challenger_door]])


        # Checking when the user wishes to change the door
        if challenger_door_changed == car_variable_door:
            wins = wins + 1
        else:
            losses = losses + 1

        win_perc = wins / (wins + losses)
        switch_doors.append(win_perc)
        # print(f"\nProbabilty of winning if the User wants to switch the door runnung {n} times:--> {win_perc}")
        # print(win_perc)
    return (win_perc),switch_doors,number_of_iterations
```

```python
result_switch_doors,switch_doors,number_of_iterations = switch_doors()
print(f"\n\nProbabilty if the User wants to switch the door::--> {result_switch_doors}")
```

```python
plt.figure(figsize = (20,6))
plt.scatter(x = number_of_iterations,y=switch_doors,alpha=0.4)
plt.xlabel("Number of iterations")
plt.ylabel("Probability of winning the car")
plt.title("Visualizing the probability values of winning the car if user does  switch the door");
```
